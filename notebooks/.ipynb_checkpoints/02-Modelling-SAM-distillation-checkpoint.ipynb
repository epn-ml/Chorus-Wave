```python
dataset = Chorus_dataset("../data/processed/", transform = True, use_sam = True)
```

```python
dataset_len = int(len(dataset))
n = 40
train_dataset, val_dataset = torch.utils.data.random_split(dataset, [n, (dataset_len - n)])
```

```python
DEVICE = torch.device("cuda:7" if torch.cuda.is_available() else "cpu")
#DEVICE = torch.device("cpu")
MODEL_TYPE = "vit_h"
CHECKPOINT_PATH = os.path.join("sam","sam_vit_h_4b8939.pth")
```

```python
sam = sam_model_registry[MODEL_TYPE](checkpoint=CHECKPOINT_PATH).to(device=DEVICE)
```

```python
from models import Decoder

decoder = Decoder().to(DEVICE)

# Set up the optimizer, hyperparameter tuning will improve performance here
lr = 1e-4
wd = 0
optimizer = torch.optim.Adam(decoder.parameters(), lr=lr, weight_decay=wd)
loss_fn = torch.nn.BCELoss()
# loss_fn = torch.nn.BCELoss()
#keys = list(bbox_coords.keys())

train_dataloader = DataLoader(train_dataset, batch_size=1, shuffle=True, num_workers=0)
val_dataloader = DataLoader(val_dataset, batch_size=1, shuffle=True, num_workers=0)
```

```python
NUM_EPOCHS = 100
losses = []

for epoch in range(NUM_EPOCHS):
  epoch_losses = []
  for i, batch in enumerate(train_dataloader):
    input_image = batch['image'].to(DEVICE)
    input_size = (640, 920)
    original_image_size = (640,920)
    
    # No grad here as we don't want to optimise the encoders
    with torch.no_grad():
        image_embedding = sam.image_encoder(input_image)
    out_mask = decoder(image_embedding)
    #binary_mask = normalize(threshold(upscaled_masks, 0.0, 0))
    ground_truth_mask = batch['mask'].to(DEVICE)
    #gt_mask_resized = torch.from_numpy(np.resize(ground_truth_mask, (1, ground_truth_mask.shape[1], ground_truth_mask.shape[2]))).to(DEVICE)
    

    loss = loss_fn(out_mask, ground_truth_mask)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    epoch_losses.append(loss.item())
  losses.append(epoch_losses)
  print(f'EPOCH: {epoch}')
  print(f'Mean loss: {mean(epoch_losses)}')
```

```python
test_dataset = Chorus_dataset("../data/processed/", split = 'test', init = True, transform = True, use_sam = True)
test_dataloader = DataLoader(test_dataset, batch_size=1, shuffle=True, num_workers=0, )
```

```python
mean_losses = [mean(x) for x in losses]
mean_losses

plt.plot(list(range(len(mean_losses))), mean_losses)
plt.title('Mean epoch loss')
plt.xlabel('Epoch Number')
plt.ylabel('Loss')

plt.show()
```

```python
def calculate_metrics(pred_mask, gt_mask):
    tp, fp, fn, tn = smp.metrics.get_stats(pred_mask, gt_mask, mode="binary", threshold=0.5)
    iou_score = smp.metrics.iou_score(tp, fp, fn, tn, reduction="micro")
    f1_score = smp.metrics.f1_score(tp, fp, fn, tn, reduction="macro")
    accuracy = smp.metrics.accuracy(tp, fp, fn, tn, reduction="macro")
    recall = smp.metrics.recall(tp, fp, fn, tn, reduction="micro-imagewise")
    precision = smp.metrics.precision(tp, fp, fn, tn, reduction="micro-imagewise")

    return iou_score, f1_score, accuracy, recall, precision


def calculate_avg_score(metrics):
    metrics = np.array(metrics)
    return np.sum(metrics, axis = 1)/len(metrics)
```
